```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = '/content/drive/MyDrive/kb차차차/최종 데이터.csv'
```

```python
df = pd.read_csv(file_path, engine = 'python', encoding = 'cp949')
```

```python
df = df.drop(columns = ['Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18'])
df
```

# 연식과 가격의 상관관계 시각화

```python
figure = plt.figure(figsize=(18, 12))
plt.rcParams.update({'font.size': 10})
sns.boxplot(x='연식', y="가격(만원)", data=df, showmeans=True,
            meanprops={"marker":"s","markerfacecolor":"black", "markeredgecolor":"black"})
plt.rc('font', family='NanumBarunGothic') 
```

```python
figure = plt.figure(figsize=(18, 12))
plt.rcParams.update({'font.size': 10})
sns.boxplot(x='연식', y="가격(만원)", hue='차종', data=df, showmeans=True,
            meanprops={"marker":"s","markerfacecolor":"black", "markeredgecolor":"black"})
plt.rc('font', family='NanumBarunGothic') 
```
