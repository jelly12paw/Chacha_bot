# 중고차 데이터

```python
file_path = '/content/drive/MyDrive/kb차차차/최종 데이터.csv'
```

```python
df = pd.read_csv(file_path, engine = 'python', encoding = 'cp949')
```

```python
df = df.drop(columns = ['Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18'])
df
```

# 원핫인코더 중고차 데이터

```python
test = pd.get_dummies(df)
test
```

# Random Forest

```python
features = test.loc[:, '연식':'변속기_오토']
X = features.values
y = test['가격(만원)']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
logreg = RandomForestRegressor(oob_score=True, n_jobs=-1, random_state=42)
logreg.fit(X_train, y_train)
print("테스트 점수 : {:.2f}".format(logreg.score(X_test,y_test)))
print("훈련셋 점수 : {:.2f}".format(logreg.score(X_train,y_train)))
```

# Gradient Boosting

```python
features = test.loc[:, '연식':'변속기_오토']
X = features.values
y = test['가격(만원)']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
grad_boost = GradientBoostingRegressor(random_state=42)
cv = cross_validate(estimator=grad_boost, X=X_train, y=y_train, n_jobs=-1, 
                    return_train_score=True)
print(np.mean(cv['train_score']))
print(np.mean(cv['test_score']))
```

# Poly(degree=3) + Standard + Linear

```python
poly = PolynomialFeatures(degree=3, include_bias=False) 
std_scaler = StandardScaler()
lin_reg = LinearRegression()
```

```python
model = Pipeline(steps=[('deg_2_poly', poly),
                        ('scaler', std_scaler),
                        ('reg', lin_reg)])
model.fit(X_train, y_train)
```

```python
train_pred = model.predict(X_train)
r2_score(y_train, train_pred)
```

```python
test_pred = model.predict(X_test)
r2_score(y_test, test_pred)
```

# 머신러닝 비교

```python
knn_clf = KNeighborsClassifier(n_jobs=-1)
log_clf = LogisticRegression(random_state=42, n_jobs=-1)
sgd_clf = SGDClassifier(loss='log', n_jobs=-1, random_state=42) 
tree_clf = DecisionTreeClassifier(random_state=42)
svm_clf = SVC(probability=True, random_state=42)
```

```python
classifiers = [knn_clf, log_clf, sgd_clf, tree_clf, svm_clf]
for clf in classifiers:
    clf.fit(X_train, y_train)  # 모델 훈련
    acc = clf.score(X_test, y_test)  # 테스트 셋 평가 점수(정확도) 계산
    print(clf.__class__.__name__, 'accuracy:', acc)
```
